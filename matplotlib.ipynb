```python
#axes,labels,legends,grid
#label is the name of each graph and to make it visible need to use legends

import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,]
max_t=[21,27,22,29,25,26]
min_t=[12,11,14,17,13,12]
avg_t= [20,21,19,19,18,22]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="max", marker="o")
plt.plot(days,min_t,label="min",marker="o")
plt.plot(days,avg_t,label="avg",marker="o")

#we create a legend to get the details of label name for each graph
#location ="best" gives the best position
#bottom right, upper left, center left are positions
#shadow =True crates a shadow


plt.legend(loc="center left",shadow=True, fontsize="large")  


#how to apply grid
plt.grid()
plt.show()   #show needs to be used
```

```python
#create new data for graph using array

import matplotlib.pyplot as plt
import numpy as np

company=["GOOGLE","AMAZON","MICROSOFT","KONECRANES","NMK"]



revenue=np.array([112,132,142,32,2])
expenses=np.array([34,64,100,27,1])
profit=revenue-expenses
profit

```

```python
#create x position value

xpos=np.arange(len(company))
xpos
```

```python
#multiple bar graph

plt.bar(xpos-0.2, revenue, width=0.4, label="revenue")
plt.bar(xpos+0.2, profit, width=0.4, label="profit")

plt.xticks(xpos, company)  #xticks is used to have name for each coloumn instead of full x axis general name
plt.ylabel("Revenue(billions)")
plt.title("US Tech Giants Stocks")
```

```python
#horizontal bar
plt.barh(xpos, revenue, label="Revenue")
plt.yticks(xpos, company)
plt.title("US Tech Stocks")
plt.legend()
```

```python
#drwa a pie chart

import matplotlib.pyplot as plt
a=[1000,500,700,300,400]
b=["Home rent","Food","car loan","Personal loan","Extra"]

plt.pie(a,labels=b)
```

```python
#draw a pie chart with perfect circle

plt.pie(a,labels=b, shadow=True)
plt.axis("equal")
```

```python
#show percentage

plt.axis("equal")
plt.pie(a,labels=b,autopct="%1.1f%%", explode=[0,0.1,0.2,0,0])
```

```python
#counter clock and angle properties

plt.axis("equal")
plt.pie(a,labels=b,autopct="%1.1f%%", counterclock=True,startangle=45)


```

```python
#how to save a chart

plt.axis("equal")
plt.pie(a,labels=b,autopct="%1.1f%%", counterclock=True,startangle=45,explode=[0,0,0.2,0,0])
plt.savefig("Piechart.jpg",bboxinches="tight", pad_inches=1)
```
