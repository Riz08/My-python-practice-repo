```python
#line plot

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[21,27,22,29,25,26]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.plot(x,y,color="green",linestyle="dotted",linewidth=5,marker="+", markersize=20 )
```

```python
#format string and other options
x=[1,2,3,4,5,6]
y=[21,27,22,29,25,26]

plt.xlabel("Day")
plt.ylabel("Tempebrature")
plt.title("Weather graph")
plt.plot(x,y,"+--",)  #+,o,& in creating the mark cross point 
```

```python
#alpha is used to control the tranparency of the chart
plt.plot(x,y,"r<",alpha=.3)  
```

```python
#axes,labels,legends,grid
#label is the name of each graph and to make it visible need to use legends

import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,]
max_t=[21,27,22,29,25,26]
min_t=[12,11,14,17,13,12]
avg_t= [20,21,19,19,18,22]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="max", marker="o")
plt.plot(days,min_t,label="min",marker="o")
plt.plot(days,avg_t,label="avg",marker="o")

#we create a legend to get the details of label name for each graph
#location ="best" gives the best position
#bottom right, upper left, center left are positions
#shadow =True crates a shadow


plt.legend(loc="center left",shadow=True, fontsize="large")  


#how to apply grid
plt.grid()
plt.show()   #show needs to be used in pycharm
```
