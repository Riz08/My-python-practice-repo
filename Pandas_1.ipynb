```python
import pandas as pd
#Read the csv file
df = pd.read_csv('nyc-weather.csv')
df.head()
```

```python
df.tail()
```

```python
#What was the max temp for Jan 2021 @NY city
df['Temperature'].max()
```

```python
#Which all dates it rained @NY for Jan 2021
df['EST'][df['Events'] == 'Rain']
```

```python
#Avg windspeed @NY for Jan 2021
df.fillna(0, inplace=True)
df['WindSpeedMPH'].mean()
```

```python
#using relational operator with df
df[df.Temperature >= 32]
```

```python
df[df.Temperature == df.Temperature.max()]
```

```python
#setting the date column as index
df.set_index('EST')
```

```python
df.set_index('Events', inplace=True)
```

```python
df.loc[['Snow','Rain',]]
```

```python
df.loc[lambda df: df['Temperature'] > 40]
```
