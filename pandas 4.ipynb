```python
#now we learn how to create a date for a data
#need to use date range method

import pandas as pd
df=pd.read_csv("appl_no_dates.csv")
df.head()
```

```python
#create a date index and replace it in the index of the table
#B stands for business days

rng=pd.date_range(start="4/1/2022",end="5/2/2022",name="Date",freq="B")
rng
```

```python
df.set_index(rng,inplace=True)
df[::-1].head() #inverse to get the latest date
```

```python
#how to find the missing dates from datetime index

daily_index=pd.date_range(start="4/1/2022",end="5/2/2022",freq="D")
daily_index
```

```python
daily_index.difference(df.index)
```

```python
#using time zone in date range

london=pd.date_range("5/14/2022 09:00:00",periods=10,freq="H",tz="Europe/London")
london
```

```python
#using custom business days to generate us holidays

from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
us_cal=CustomBusinessDay(calendar=USFederalHolidayCalendar())


rng=pd.date_range(start="7/1/2022",end="8/02/2022",freq=us_cal)
rng
```

```python
df.set_index(rng,inplace=True)
df.tail()
```
