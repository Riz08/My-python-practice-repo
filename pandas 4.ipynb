```python
#now we learn how to create a date for a data
#need to use date range method

import pandas as pd
df=pd.read_csv("appl_no_dates.csv")
df.head()
```

```python
#create a date index and replace it in the index of the table
#B stands for business days

rng=pd.date_range(start="4/1/2022",end="5/2/2022",name="Date",freq="B")
rng
```

```python
df.set_index(rng,inplace=True)
df[::-1].head() #inverse to get the latest date
```
