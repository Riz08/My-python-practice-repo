```python
#pandas concat
import pandas as pd
india_weather = pd.DataFrame({
    "city" : ['Mumbai', 'Delhi', 'Chennai'],
    "temperature" : [37, 43, 41],
    "humidity" : [85, 60, 83]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city' : ['Newyork', 'Chicago', 'Orlando'],
    'temperature' : [21, 14, 35],
    'humidity' : [68, 65, 75]
})
us_weather
```

```python
from re import U
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
df.to_csv('concat.csv', index_label=None)
```

```python
#pandas merge
import pandas as pd
df1 = pd.DataFrame({
    "city" : ["newyork", "chicago","orlando"],
    "temperature" : [21, 14, 35]
})
```

```python
df2 = pd.DataFrame({
    "city" : ['chicago','newyork','orlando'],
    "humidity":[65, 68, 75]
})
```

```python
df = pd.merge(df1,df2, on="city")
df
```

```python
df1 = pd.DataFrame({
    "city" : ['newyork','chicago','orlando','baltimore'],
    "temperature" : [21, 14, 35, 38]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ['chicago','newyork','loseangels'],
    "humidity" : [65, 68, 70]
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city', how='inner')
df3
```

```python
df3 = pd.merge(df1, df2, on='city', how='outer')
df3
```

```python
df3 = pd.merge(df1, df2, on='city', how='left')
df3
```

```python
df3 = pd.merge(df1, df2, on='city',how='right')
df3
```

```python
df3 = pd.merge(df1,df2, on='city',how='outer',indicator=True)
df3
```
